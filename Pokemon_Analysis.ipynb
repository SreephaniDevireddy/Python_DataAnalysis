```python
import pandas as pd
```

```python
df = pd.read_csv("../Pandas/Pokemon.csv")
```

```python
df
```

```python
df.dtypes
```

```python
#changing datatypes using dictonary
change_type = {'#':str}

df = df.astype(change_type)

df.dtypes
```

```python
df['Type 2'].head(30)
df['Type 2'].tail(30)
```

<H4> filling those NaN values with "None"

```python
df['Type 2'] = df['Type 2'].fillna('None')
df
```

<H1> Modifying the Dataset 

```python
#changing the column name using dictionary
df_pk = df.rename(columns={"#":"PK_ID", "Type 1":"Type1", "Type 2":"Type2"})

#dropping the column - Total
df_pk = df_pk.drop(columns = ['Total'])
```

```python
#Adding another column as Total (again using the iloc method)
#iloc method explaination for the below - including all rows and column from 4 to 9

df_pk['Total'] = df_pk.iloc[:,4:10].sum(axis=1)   #here axis=1, to add it as a column (vertically)
df_pk
```

<H2>Logic I followed in this dataset is to split into 2 dfs 
  <h4>  1. Pokemons which are both types
  <h4>  2. PK which contains only one type

```python
df_pk.loc[df_pk['Type2'] != 'None'].count()
```

```python
#PKs which has both types
df_both = df_pk.loc[df_pk['Type2'] != 'None']
df_both.reset_index(drop=True, inplace = True)
df_both
```

```python
#PKs which have only one type
df_one = df_pk.loc[df_pk['Type2'] == 'None']
df_one.reset_index(drop=True, inplace = True) #to reset the index to the same df we use inplace (to get the same df from memory)
df_one
```

<h2> EDA on both dfs = df_both & df_one

<h3>EDA on df_both

```python
df_both.groupby(['Type1']).sum().sort_values('Total',ascending=False)
```

```python
#what are the unique values for Type2 PKs
df_both.Type2.unique()
```

#Combining both Type1/Type2 to create a new column in the df_both, which can make analysis easiler further

```python
df_both['Type'] = df_both['Type1'] + '/' + df_both['Type2']
df_both
```

```python
#stats with new column - Type
df_both.groupby(['Type']).sum().sort_values('Total', ascending = False)
```

What are the number of different PKs with dual type per Generation 

```python
df_both.groupby('Generation').nunique().sort_values('Type',ascending=False)
```

<h3>EDA on df_one

```python
#basic stats for single type PKs
df_one.groupby(['Type1']).sum().sort_values('Total', ascending = False)
```

What are the number of different PKs per Generation 

```python
df_one.groupby('Generation').nunique().sort_values('Type1',ascending=False)
```

Inferences - Based on the above we can infer that df_both and df_one have different stats 
Further the analysis can be done separately to get better results on the attack mechanism

<h3> NORMAL ATTACKS - VISUALIZATION

```python
fig = plt.gcf() 
# Changing Seaborn Plot size
fig.set_size_inches(10, 7)


sns.barplot(x='Generation',y='Defense',
               hue = 'Legendary',
           data = df_both)

plt.legend(loc='upper right', title = 'Legendary',borderaxespad=0)
plt.title('')
```

Inference -
The scatterplots for each Generation between HitPoints and Defense says that Generation 2 PKs have more highest and consistent distribution values and therefore are more resistant to NORMAL ATTACKS

Question 1 - What are the top 10 Pokemons that can do highest damage in an ATTACK

```python
df_2 = df_both.groupby('Type')['Total'].sum().nlargest(10)
```

```python
df_2.plot(kind='bar', title='Top 10 Pokemons with Highest Damage possible in an Attack',
               ylabel='', xlabel='Type', figsize=(20, 6))

#rotate x-axis ticks vertically
plt.xticks(rotation=0)
```

Question 2 - Which Generations can do more damage

```python
df_3 = df_both.groupby(['Type','Generation'])['Total'].sum().nlargest(10)

fig = plt.gcf() 
# Changing Seaborn Plot size
fig.set_size_inches(15, 5)

df_3 = df_3.reset_index()
# plot barplot
sns.barplot(x="Type",
           y="Total",
            hue='Generation',
           data=df_3
           ).set(title='Top 10 Pokemons - Generation wise')
```

Question 3 - What are the Generations for the TOP 10 Pokemons with highest Speed

```python
df_4 = df_both.groupby(['Type','Generation'])['Speed'].sum().nlargest(10)

fig = plt.gcf() 
# Changing Seaborn Plot size
fig.set_size_inches(15, 5)

df_4 = df_4.reset_index()
# plot barplot
sns.barplot(x="Type",
           y="Speed",
            hue='Generation',
           data=df_4
           ).set(title='Top 10 Pokemons with Highest Speed- Generation wise')
```

Inference - Generation 1 PKs have highest speed
